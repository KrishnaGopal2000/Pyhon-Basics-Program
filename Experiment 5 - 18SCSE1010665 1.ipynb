## 2.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  
Y = iris.target
```

```python
logreg = LogisticRegression(C=1e5)
```

```python
logreg.fit(X, Y)
```

```python
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.show()
```

## 3.

```python
X = iris.data[:, -2:]  
Y = iris.target
```

```python
logreg = LogisticRegression(C=1e5)

logreg.fit(X, Y)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .01
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
```

According to the model, these flowers have been clsssified into 3 different classes. Please find below a summary of the characteristics for each of them:
* *__Class1(Color Blue)__*: These are the flowers with a shorter petal length and width with a range of 1-2 cm petal length and 0.5 cm petal width
* *__Class2(Color Orange)__*: These are medium sized flowers, with average petal length of 4 cm and width between 1 and 1.5 cm
* *__Class3(Color Brown)__*: These are big flowers with petal length generally more than 4.5 cms and petal width more than 1.5 cm
